# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
x = np.array([1,2,3,4,5])
```

```python
x
```

```python
x_sq = x**2
```

```python
x_sq
```

```python
# plt.plot(x,x_sq , marker = "*")
plt.plot(x,x_sq,marker = "o",label = "y = x^2")
plt.plot(x,x,marker = "*" , label = "y = x")
plt.xlabel("x")
plt.ylabel("x_sq")
plt.title("Comparison function")
plt.legend() # it displays the labels
# plt.xlim(-10,10)
plt.show()
```

```python
plt.plot(x,x,marker = "o",color = "red")
plt.show()
```

### Scatter Plot

```python
plt.scatter(x,x_sq,marker = "o",label = "y = x^2")
plt.scatter(x,x,marker = "*" , label = "y = x")
plt.xlabel("x")
plt.ylabel("x_sq")
plt.title("Comparison function")
plt.legend() # it displays the labels
# plt.xlim(-10,10)
plt.show()
```

### Bar Graph

```python
x = np.array([1,2,3,4,5])
sales_in_2018 = np.array([5,3,1,8,9])
sales_in_2019 = np.array([2,6,9,1,7])
```

```python
plt.bar(x,sales_in_2018,width=0.3)
# plt.bar(x,sales_in_2019,width=0.3,alpha = 0.5)
plt.bar(x + 0.3,sales_in_2019,width=0.3)
plt.xlabel("Car numbers")
plt.ylabel("Sales")
plt.show()
```

```python
plt.bar(x,sales_in_2019)
plt.show()
```
