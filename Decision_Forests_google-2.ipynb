```python
import ydf # Yggdrasil Decision Forest
import numpy as np
import pandas as pd
```

```python
# The Palmer Penguins dataset
```

```python
# Classification problem
```

```python
path = "https://storage.googleapis.com/download.tensorflow.org/data/palmer_penguins/penguins.csv"
dataset = pd.read_csv(path)
```

```python
label = "species"
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
dataset.isna().sum()
```

```python
# Unlike neural networks , decision forests support all these feature types natively , so you don't have to do one-hot encoding , normalization , or extra is_present feature.
```

```python
np.random.seed(1)
is_test = np.random.rand(len(dataset)) < 0.2

train_dataset = dataset[~is_test]
test_dataset = dataset[is_test]

print("Training examples :" , len(train_dataset))
print("Testing examples : ",len(test_dataset))
```

```python
len(dataset)
```

```python
dataset.shape
```

```python
# Training a decision trees with default hyperparameters
```

```python
# You can train your first decision tree with the CART(Classification and Regression Trees) learning algorithm (a.k.a. learner) without specifying any hyperparameters.
```

```python
# That's because the ydf.CartLearner learner provides good default hyperparameter values.
```

```python
model = ydf.CartLearner(label = label).train(train_dataset)
```

```python
model.plot_tree()
```

```python
train_evaluation = model.evaluate(train_dataset)
print("train accuracy : ", train_evaluation.accuracy)
```

```python
test_evaluation = model.evaluate(test_dataset)
print("test accuracy : ",test_evaluation.accuracy)
```

```python
# train and test was split randomly
```

```python
# for such a small dataset , using cross-validation would be preferable as it would compute more accurate evolution metric values.
```

```python
# Improving the model hyperparameters
```

```python
model = ydf.CartLearner(label = label,min_examples=1).train(train_dataset)
model.plot_tree()
```

```python
print(model.evaluate(test_dataset).accuracy)
```

```python
# The quality of the model is decreased
```

```python
model = ydf.RandomForestLearner(label = label).train(train_dataset)
print("Test accuracy : ",model.evaluate(test_dataset).accuracy)
```

```python
# accuracy of the random forest is better than our simple tree.
```
