```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(n_samples=500,n_features=2,centers= 3,random_state=3)
```

```python
x.shape
```

```python
y.shape
```

```python
x[:5]
```

```python
y[:10]
```

```python
plt.scatter(x[:,0],x[:,1],c= y)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(x)
```

```python
locations = model.cluster_centers_
locations
```

```python
plt.scatter(x[:,0],x[:,1])
plt.scatter(locations[:,0],locations[:,1],s = 300 , marker = "*")
plt.show()
```

```python
predicted_labels = model.labels_
predicted_labels
```

```python
plt.scatter(x[:,0],x[:,1],c = predicted_labels)
plt.scatter(locations[:,0],locations[:,1],s = 300,marker = "*")
plt.show()
```

### Dominant Color Extraction from an image

```python
img = plt.imread("tree.jpg")
```

```python
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
n = img.shape[0]*img.shape[1]
n
```

```python
# flattening an image
all_pixels = img.reshape((n,3))
all_pixels.shape
```

```python
all_pixels[:5]
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(all_pixels)
```

```python
centers = model.cluster_centers_.astype('uint8')
centers
```

#### Regenerate our image using only top 3 colors.

```python
new_img = np.zeros((n,3),dtype='uint8')
new_img.shape
```

```python
centers
```

```python
model.labels_
```

```python
for i in range(n):
    group_idx = model.labels_[i]
    new_img[i] = centers[group_idx]
```

```python
new_img.shape
```

```python
new_img = new_img.reshape(*img.shape)
```

```python
plt.imshow(new_img)
plt.show()
```
