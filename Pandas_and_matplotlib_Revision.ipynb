## Matplotlib

```python
import numpy as np
```

```python
import pandas as pd
```

```python
x = np.array([1,2,3,4,5])
```

```python
x
```

```python
x_sq = x**2
```

```python
x_sq
```

```python
plt.plot(x,x_sq,marker = "o",label="y = x^2")
plt.plot(x,x,marker = "*",label = "y = x")
plt.xlabel("x")
plt.ylabel("x_sq")
plt.title("Comparison function")
plt.legend()
# plt.xlim(-10,10)
plt.show()
```

```python
plt.plot(x,x,marker = "o",color = "red")
plt.show()
```

### BarGraph

```python
x = np.array([1,2,3,4,5])
sales_in_2018 = np.array([5,3,1,8,9])
sales_in_2019 = np.array([2,6,9,1,7])
```

```python
plt.bar(x,sales_in_2018,width = 0.3)
# plt.bar(x,sales_in_2019,width=0.3,alpha = 0.5)
plt.bar(x+0.3,sales_in_2019,width=0.3)
plt.xlabel("Car numbers")
plt.ylabel("Sales")
plt.show()
```

```python
plt.bar(x,sales_in_2019)
plt.show()
```

## OpenCV - open source computer vision library

```python
import cv2
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread("group.jpg")
```

```python
img
```

```python
type(img)
```

```python
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
all_faces = detector.detectMultiScale(img)
```

```python
all_faces
```

```python
for one_face in all_faces:
    x,y,w,h = one_face
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,0),2)
```

```python
plt.imshow(img)
plt.show()
```
