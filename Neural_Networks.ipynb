```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons , make_circles , make_classification
```

```python
def load_dataset(dataset):
    if dataset == 'moons':
        X,Y = make_moons(n_samples=10000 , noise = 0.07 , random_state=1)
    elif dataset == 'circles':
        X,Y = make_circles(n_samples=10000 , noise = 0.07 , random_state= 1 , factor=0.5)
    elif dataset == 'classification':
        X,Y = make_classification(n_samples=10000 , n_features = 2 , n_redundant=0 , random_state=2)
    return X,Y
```

```python
dataset = 'circles'
X,Y = load_dataset(dataset)
print(X.shape , Y.shape)
```

```python
plt.scatter(X[:,0] , X[:,1],c = Y)
plt.show()
```

### Model Building - ANN

```python
from tensorflow.keras import Sequential # we want to create sequential model
```

```python
from tensorflow.keras.layers import Dense # fully connected layer
```

```python
model = Sequential()
```

```python
model.add(Dense(5,activation = 'relu',input_shape = (2,))) # 5 is number of neurons and input_shape is only applied for first layer
model.add(Dense(5,activation = 'relu'))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy',metrics = ['accuracy']) # it will tell the model that we wants to do binary classification with neural networks
```

```python
model.fit(X,Y,batch_size=64,epochs = 50, validation_split=0.2)
```
